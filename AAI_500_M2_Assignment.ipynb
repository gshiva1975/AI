Data analysts often implement statistical inference methods by setting the probability of a
correct inference equal to 0.95. Let *A* denote the event that an inference for the population
about men is correct. Let *B* represent the event of a corresponding inference about women
being correct. Suppose that these are independent events.


(a) Find the probability that (i) *both* inferences are correct, (ii) *neither* inference is correct.  
(b) Construct the probability distribution for *Y* = number of correct inferences.  
(c) With what probability would each inference need to be correct in order for the probability to be 0.95 that *both* are correct?



```python
# Answer
import math

# Given
prob_A_correct = 0.95
prob_B_correct = 0.95

# Part (a) Calculations P(A and B) = P(A)*P(B)
# (i) Probability that both inferences are correct
prob_both_correct = prob_A_correct * prob_B_correct

# (ii) Probability that neither inference is correct P(not A) = 1 - P(A) , P(not B) = 1 - P(B)
prob_A_incorrect = 1 - prob_A_correct
prob_B_incorrect = 1 - prob_B_correct
prob_neither_correct = prob_A_incorrect * prob_B_incorrect

# Part (b) Constructing the probability distribution for Y
# Probability that exactly one inference is correct
prob_one_correct = (prob_A_correct * prob_B_incorrect) + (prob_A_incorrect * prob_B_correct)

# Probability distribution for Y = 0, 1, 2
prob_distribution = {
    'Y = 0 (Neither correct)': prob_neither_correct,
    'Y = 1 (One correct)': prob_one_correct,
    'Y = 2 (Both correct)': prob_both_correct
}

# Part (c) Finding the probability needed for both to be correct with 0.95 probability
# P(Bothcorrect)=pp=p square
# p ^2 =0.95
# p= 0.950.97

desired_prob_both_correct = 0.95
prob_each_needed = math.sqrt(desired_prob_both_correct)

# Printing the results
print("Part (a):")
print(f"Probability that both inferences are correct: {prob_both_correct:.2f}")
print(f"Probability that neither inference is correct: {prob_neither_correct:.2f}")

print("\nPart (b): Probability Distribution for Y:")
for outcome, prob in prob_distribution.items():
    print(f"{outcome}: {prob:.2f}")

print("\nPart (c):")
print(f"The probability needed for each inference to be correct for both to be correct with probability 0.95: {prob_each_needed:.2f}")
```

## Problem # 2.21.

Plot the gamma distribution by fixing the shape parameter *k* = 3 and setting the scale parameter
= 0.5, 1, 2, 3, 4, 5. What is the effect of increasing the scale parameter? (See also Exercise 2.48.)

```python
# Answer
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Fixed shape parameter k
k = 3

# Different scale parameters
scale_parameters = [0.5, 1, 2, 3, 4, 5]

# Generate x values for plotting
x_values = np.linspace(0, 20, 1000)

# Plot the gamma distributions with different scale parameters
plt.figure(figsize=(12, 8))

for scale in scale_parameters:
    # Calculate the gamma distribution pdf values
    pdf_values = gamma.pdf(x_values, k, scale=scale)
    # Plot each distribution
    plt.plot(x_values, pdf_values, label=f'Scale = {scale}')

# Adding labels and title
plt.title(f'Gamma Distribution (Shape parameter k = {k}) with Varying Scale Parameters')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

The effect of increasing the scale parameter (

) in the gamma distribution can be described as follows:

Spread of the Distribution: As the scale parameter increases, the distribution becomes more spread out, indicating a higher variance.
Shift of the Peak: The peak (or mode) of the distribution shifts to the right.
Height of the Peak: The peak's height decreases because the distribution becomes more spread out over a larger range of values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
sensitivity = 0.999  # Sensitivity of the test
specificity = 0.9999  # Specificity of the test
prevalence = 1 / 10000  # Prevalence of HIV (1 in 10,000 men)

# Function to calculate the Positive Predictive Value (PPV)
def calculate_ppv(prevalence, sensitivity, specificity):
    ppv = (sensitivity * prevalence) / ((sensitivity * prevalence) + ((1 - specificity) * (1 - prevalence)))
    return ppv

# Calculate PPV for the given prevalence
ppv = calculate_ppv(prevalence, sensitivity, specificity)

# Explanation of the results
print(f"Sensitivity: {sensitivity} (99.9%) - The probability of correctly identifying an HIV-positive individual.")
print(f"Specificity: {specificity} (99.99%) - The probability of correctly identifying an HIV-negative individual.")
print(f"Prevalence: {prevalence} - The proportion of the population that is actually HIV-positive.")
print(f"Calculated Positive Predictive Value (PPV): {ppv:.4f} or {ppv*100:.2f}%")
print("\nExplanation:")
print(f"Despite the high sensitivity and specificity of the test, the PPV is only about {ppv*100:.2f}%.")
print("This means that if a man tests positive, there is only around a 50% chance he actually has HIV.")
print("The low PPV occurs because of the low prevalence of the disease, leading to many false positives.")

# Plotting the PPV as a function of prevalence
prevalence_range = np.logspace(-7, -1, 100)  # Prevalence from 1 in 10 million to 1 in 10
ppv_values = [calculate_ppv(p, sensitivity, specificity) for p in prevalence_range]

plt.figure(figsize=(10, 6))
plt.plot(prevalence_range, ppv_values, label='Positive Predictive Value (PPV)', color='blue')
plt.axvline(x=prevalence, color='red', linestyle='--', label=f'Prevalence: 1 in 10,000 (HIV)')
plt.xscale('log')  # Logarithmic scale for prevalence
plt.xlabel('Prevalence (log scale)')
plt.ylabel('Positive Predictive Value (PPV)')
plt.title('Positive Predictive Value as a Function of Disease Prevalence')
plt.legend()
plt.grid(True)
plt.show()
```

HIV Test Example: The plot demonstrates that the PPV is very low when the prevalence of the disease is rare, indicating a high rate of false positives. This underscores the potential disadvantage of routine diagnostic screening for rare diseases: even highly accurate tests can produce many false positives.

Importance of Targeted Testing: The example shows that routine screening for rare diseases in the general population may not always be beneficial. Instead, targeted testing (screening populations with higher prevalence or risk factors) can be more effective in ensuring a higher PPV.

## Problem # 2.27.

The distribution of *X* = heights *(cm)* of women in the U.K. is approximately *N*(162, $7^2$).
Conditional on *X = x*, suppose *Y* = weight *(kg)* has a *N*(3.0 + 0.40x, $8^2$) distribution. Simulate
and plot 1000 observations from this approximate bivariate normal distribution. Approximate
the marginal means and standard deviations for *X* and *Y*. Approximate and interpret the
correlation.


```python
import numpy as np
import matplotlib.pyplot as plt

# Giveb mean = 162 , variance

# Parameters for the normal distributions
mean_X = 162  # Mean of heights
std_dev_X = np.sqrt(7*7)  # Standard deviation of heights (sqrt of variance)

# Conditional distribution parameters for Y given X
intercept_Y = 3.0
slope_Y = 0.40
std_dev_Y = np.sqrt(8*8)  # Standard deviation of weights (sqrt of variance)

# Number of observations to simulate
num_samples = 1000

# Simulate heights (X) from N(162, 7^2)
X = np.random.normal(mean_X, std_dev_X, num_samples)

# Simulate weights (Y) from N(3.0 + 0.40 * X, 8^2)
Y = np.random.normal(intercept_Y + slope_Y * X, std_dev_Y, num_samples)

# Plot the simulated bivariate normal distribution
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, alpha=0.75, edgecolors='k', linewidths=0.5)
plt.title('Simulated Bivariate Normal Distribution of Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()

# Calculate and approximate marginal means and standard deviations
mean_X_approx = np.mean(X)
std_dev_X_approx = np.std(X)
mean_Y_approx = np.mean(Y)
std_dev_Y_approx = np.std(Y)

# Calculate the correlation between X and Y
correlation_approx = np.corrcoef(X, Y)[0, 1]

print(f"{mean_X_approx:.2f}")
print(f"{std_dev_X_approx:.2f}")
print(f"{mean_Y_approx:.2f}")
print(f"{std_dev_Y_approx:.2f}")
print(f"{correlation_approx:.2f}")
```

Results:
Approximate mean of height: 161.84 cm

Approximate standard deviation of height: 8.62 cm

Approximate mean of weight: 67.99 kg

Approximate standard deviation of weight: 9.61 kg

Approximate correlation: 0.34

**Interpretation:**
The correlation of approximately 0.34 indicates a moderate positive relationship between height and weight in this simulated data. As height increases, weight tends to increase, but the relationship is not perfectly linear, implying other factors may influence weight as well
