In an exit poll of 1648 voters in the 2020 Senatorial election in Arizona, 51.5% said they voted
for Mark Kelly and 48.5% said they voted for Martha McSally

a) Suppose that actually 50% of the population voted for Kelly. If this exit poll had the
properties of a simple random sample, find the standard error of the sample proportion
voting for him.

b) Under the 50% presumption, are the results of the exit poll surprising? Why? Would you
be willing to predict the election outcome? Explain by (i) conducting a simulation; (ii)

answer (a)

```python
import numpy as np
import matplotlib.pyplot as plt

# standard error/std deviation calculation
def standard_error_deviation(p, n):
    return np.sqrt((p * (1 - p)) / n)

# Range of sample sizes (from 100 to 5000, with step size of 50)
sample_sizes = np.arange(100, 5000,50)

# Population proportion (50% for Mark Kelly)
p = 0.50

# Calculate SE
se = standard_error_deviation(p, 1648)

# Output the result
print(f"Standard Error (SE) of the sample proportion: {se:.4f}")
# Calculate SE for each sample size
se_values = [standard_error_deviation(p, n) for n in sample_sizes]

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, se_values, label="Standard Error (SE)")
plt.xlabel("Sample Size (n)")
plt.ylabel("Standard Error (SE)")
plt.title("Standard Error vs. Sample Size for p = 0.50")
plt.grid(True)
plt.legend()
plt.show()
```

## Problem 3.3.

 The 49 students in a class at the University of Florida made blinded evaluations of pairs of
cola drinks. For the 49 comparisons of Coke and Pepsi, Coke was preferred 29 times. In the
population that this sample represents, is this strong evidence that a majority prefers Coke?
Use a simulation of a sampling distribution to answer.

answer

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_students = 49  # Number of students
n_prefers_coke = 29  # Number of students who preferred Coke in the sample
p_null = 0.5  # Null hypothesis proportion (no preference)

# (i) Simulate the sampling distribution under the null hypothesis
n_simulations = 10000  # Number of simulations
print(n_simulations)
simulated_coke_preferences = np.random.binomial(n_students, p_null, n_simulations)

# (ii) Calculate the proportion of simulations where 29 or more students prefer Coke
p_value = np.mean(simulated_coke_preferences >= n_prefers_coke)

# Plot the simulated distribution
plt.figure(figsize=(10, 6))
plt.hist(simulated_coke_preferences, bins=np.arange(20, 50) - 0.5, color='blue', edgecolor='black', alpha=0.7)
plt.axvline(n_prefers_coke, color='red', linestyle='dashed', linewidth=2, label=f'Observed Coke Preferences = {n_prefers_coke}')
plt.title('Simulated Sampling Distribution for Coke Preferences (Under Null Hypothesis)')
plt.xlabel('Number of Students Preferring Coke')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Print p-value
print(f"P-Value: {p_value:.4f}")

# Interpretation of the p-value
if p_value < 0.05:
    print("The result is statistically significant. There is strong evidence that the majority prefers Coke.")
else:
    print("The result is not statistically significant. There is no strong evidence that the majority prefers Coke.")
```

b) In the first scenario, we chose $\sigma = 5$ under the belief that if $\mu = 20$, about 2/3 of the sample values would fall between `$`15 and `$`25. For the gamma distribution with $(\mu, \sigma) = (20,5),$ show that the actual probability between 15 and 25 is 0.688.

`Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Given parameters
mu = 20  # Mean of the gamma distribution
sigma = 5  # Standard deviation of the gamma distribution

# Step 1: Calculate shape (k) and rate (lambda)
k = (mu / sigma) ** 2  # Shape parameter
theta = sigma ** 2 / mu  # Scale parameter (theta = 1 / lambda)

# Step 2: Calculate the probability that a value is between 15 and 25
probability_between_15_and_25 = gamma.cdf(25, k, scale=theta) - gamma.cdf(15, k, scale=theta)

# Step 3: Plot the Gamma distribution and shade the area between 15 and 25
x = np.linspace(0, 40, 1000)
y = gamma.pdf(x, k, scale=theta)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Gamma Distribution (mean={mu}, sigma={sigma})', color='blue')

# Shade the area between 15 and 25
x_fill = np.linspace(15, 25, 1000)
y_fill = gamma.pdf(x_fill, k, scale=theta)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.3, label='Probability between 15 and 25')

plt.title(f'Gamma Distribution with μ={mu}, σ={sigma}')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.axvline(15, color='black', linestyle='dashed')
plt.axvline(25, color='black', linestyle='dashed')
plt.legend()
plt.grid(True)
plt.show()

# Output the probability
print(f"Probability that the value is between 15 and 25: {probability_between_15_and_25:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simulate_coin_flips(n_flips, n_trials):
    """Simulate n_trials coin flips of n_flips size and return the proportions of heads."""
    results = np.random.binomial(n_flips, 0.5, n_trials)
    proportions_of_heads = results / n_flips
    return proportions_of_heads

# Parameters
n_trials = 10000  # Number of simulations
flip_sizes = [1, 2, 3, 4]  # Different flip sizes to simulate

# Plot the distributions of proportions of heads for different flip sizes
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, n_flips in enumerate(flip_sizes):
    # Simulate the coin flips
    proportions = simulate_coin_flips(n_flips, n_trials)

    # Plot histogram of the proportions
    row = i // 2
    col = i % 2
    axs[row, col].hist(proportions, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

    # Plot the normal distribution curve (for large n)
    mean, std_dev = 0.5, np.sqrt(0.25 / n_flips)
    x = np.linspace(0, 1, 1000)
    y = norm.pdf(x, mean, std_dev)
    axs[row, col].plot(x, y, 'r-', lw=2, label=f'Normal Approximation\nμ=0.5, σ={std_dev:.4f}')

    # Titles and labels
    axs[row, col].set_title(f'Sampling Distribution (n={n_flips} flips)')
    axs[row, col].set_xlabel('Proportion of Heads')
    axs[row, col].set_ylabel('Density')
    axs[row, col].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simulate_sampling_uniform(n, n_samples):
    """Simulate random sampling from a uniform distribution and calculate the sample means."""
    # Generate n_samples sets of n uniformly distributed values between 0 and 1
    samples = np.random.uniform(0, 1, (n_samples, n))

    # Calculate the mean of each sample
    sample_means = np.mean(samples, axis=1)
    return sample_means

# Parameters
n_samples = 10000  # Number of samples
sample_sizes = [5, 30, 100]  # Different sample sizes

# Plot the distributions of sample means for different sample sizes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, n in enumerate(sample_sizes):
    # Simulate the sample means
    sample_means = simulate_sampling_uniform(n, n_samples)

    # Plot histogram of the sample means
    axs[i].hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

    # Plot the normal distribution curve for reference
    mean, std_dev = 0.5, np.sqrt(1 / (12 * n))  # Mean and standard deviation of the uniform distribution sample means
    x = np.linspace(0, 1, 1000)
    y = norm.pdf(x, mean, std_dev)
    axs[i].plot(x, y, 'r-', lw=2, label=f'Normal Approximation\nμ=0.5, σ={std_dev:.4f}')

    # Titles and labels
    axs[i].set_title(f'Sample Size n={n}')
    axs[i].set_xlabel('Sample Mean')
    axs[i].set_ylabel('Density')
    axs[i].legend()

plt.tight_layout()
plt.show()
```

As the sample size increases, the sampling distribution of the sample means becomes more normal and more concentrated around the true population mean (in this case, 0.5 for a uniform distribution).
The variance of the sampling distribution decreases as
𝑛
n increases, which means that with larger samples, the sample mean is a more precise estimate of the true population mean.
Even though the population distribution is uniform (not normal), the Central Limit Theorem ensures that the distribution of the sample means converges to a normal distribution as the sample size grows. This is a key result in statistics, allowing us to make inferences about populations based on sample means.

## Problem 3.26.

When sample data were used to rank states by brain cancer rates, Ellenberg (2014) noted
that the highest ranking state (South Dakota) and the nearly lowest ranking state (North
Dakota) had relatively small sample sizes. Also, when schools in North Carolina were ranked
by their average improvement in test scores, the best and the worst schools were very small
schools. Explain how these results could merely reflect how the variability of sample means and
proportions depends on the sample size.

`Your answer here`

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for simulation
true_improvement_rate = 0.05  # True improvement rate in test scores (5% improvement)
small_school_size = 30  # Small school size
large_school_size = 1000  # Large school size
iterations = 10000  # Number of simulations

# Simulate small school test score improvements
small_school_improvement = np.random.normal(true_improvement_rate, 0.1, (iterations, small_school_size)).mean(axis=1)

# Simulate large school test score improvements
large_school_improvement = np.random.normal(true_improvement_rate, 0.1, (iterations, large_school_size)).mean(axis=1)

# Plot the distributions for small and large schools
plt.figure(figsize=(12, 6))

# Small school size plot
plt.subplot(1, 2, 1)
plt.hist(small_school_improvement, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title('Small School Size (n=30)')
plt.xlabel('Average Test Score Improvement')
plt.ylabel('Frequency')

# Large school size plot
plt.subplot(1, 2, 2)
plt.hist(large_school_improvement, bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Large School Size (n=1000)')
plt.xlabel('Average Test Score Improvement')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Calculate and print the standard deviations (variability) of both distributions
small_school_std = np.std(small_school_improvement)
large_school_std = np.std(large_school_improvement)

print(f"Standard deviation (variability) with small school size (n=30): {small_school_std:.6f}")
print(f"Standard deviation (variability) with large school size (n=1000): {large_school_std:.6f}")
```

Small School Simulation: We simulate test score improvements for a small school of 30 students. The results show higher variability in average test score improvements due to the smaller sample size.
Large School Simulation: We simulate the same test score improvements for a large school with 1000 students. The larger sample size results in less variability in the average improvement.


Small Schools (30 students) exhibit a higher standard deviation in test score improvements, meaning that their scores are more likely to be extreme (either very high or very low) just due to chance.
Large Schools (1000 students) have a much smaller standard deviation, meaning their average improvements are more stable and less likely to show extreme values by chance.
This reflects the ranking pattern seen in North Carolina, where small schools often ranked either best or worst due to the high variability in their small sample sizes.
