```python
import tensorflow as tf
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Reshape the data for the decision tree
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)

# Create a decision tree classifier
tree_classifier = DecisionTreeClassifier(max_depth=3)  # Limit depth for visualization

# Train the model
tree_classifier.fit(x_train, y_train)

# Make predictions and evaluate
y_pred = tree_classifier.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(tree_classifier, filled=True, feature_names=range(x_train.shape[1]), class_names=[str(i) for i in range(10)])
plt.show()
```
